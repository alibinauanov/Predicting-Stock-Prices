```python
def visualize_stock_trend(figsize, values, column_name, add_moving_avg=False, moving_avg_window=30, color='blue'):
    plt.figure()
    values.plot(figsize=figsize, color=color)

    # optionally, plot a moving average
    if add_moving_avg:
        moving_avg = values.rolling(window=moving_avg_window).mean()
        plt.plot(moving_avg, label=f'{moving_avg_window}-day MA', color='red', linestyle='--')

    plt.xlabel("Years")
    plt.ylabel(column_name)
    plt.title(f"{column_name} of Google data")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
for column in google_data.columns:
  visualize_stock_trend((15,5), google_data[column], column, add_moving_avg=True, moving_avg_window=50)
```

```python
Adj_close_price = google_data[['Adj Close']]
```

```python
max(Adj_close_price.values),min(Adj_close_price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
len(scaled_data)
```

```python
x_data = []
y_data = []

for i in range(100, len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

import numpy as np
x_data, y_data = np.array(x_data), np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
int(len(x_data)*0.7)
```

```python
4908-100-int(len(x_data)*0.7)
```

```python
split_len = int(len(x_data)*0.7)
x_train = x_data[:split_len]
y_train = y_data[:split_len]

x_test = x_data[split_len:]
y_test = y_data[split_len:]
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()
# every layer even slayer
# next layer decreases the number memory cells, then it makes a decision
# to choose one cell. it's kind of filter
model.add(LSTM(128, return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(LSTM(64,return_sequences=False))
model.add(Dense(25))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, batch_size=1, epochs = 2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
inv_predictions = scaler.inverse_transform(predictions)
inv_predictions
```

```python
inv_y_test = scaler.inverse_transform(y_test)
inv_y_test
```

```python
rmse = np.sqrt(np.mean( (inv_predictions - inv_y_test)**2))
```

```python
rmse
```

```python
ploting_data = pd.DataFrame(
 {
  'original_test_data': inv_y_test.reshape(-1),
    'predictions': inv_predictions.reshape(-1)
 } ,
    index = google_data.index[split_len+100:]
)
ploting_data.head()
```

```python
visualize_stock_trend((15,6), ploting_data, 'test data')
```

```python
visualize_stock_trend((15,6), pd.concat([Adj_close_price[:split_len+100],ploting_data], axis=0), 'whole data')
```
